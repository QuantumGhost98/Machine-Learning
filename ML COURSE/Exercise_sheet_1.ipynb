### 1.5 Sampling from a Probabilistic Model

*Topic: frequentist probabilities*

Let's say we have a probabilistic model with

| X | p(X) |
|------|------|
| Sun (0)   | 0.2 |
| Rain (1)  | 0.3 |
| Snow (2)  | 0.5 |

a. Write a sampler that randomly samples from $p(X)$ values X = 0, 1, 2. In your solution, you are allowed to use the function `numpy.random.random()` but other than this, do not use any other imported functions.

```python
import numpy as np
import matplotlib.pyplot as plt

def return_random_sample():
    value = 0
    random_value = np.random.random()
    if random_value < 0.2:
        value = 0
    elif random_value <0.5:
        value = 1
    elif random_value <1:
        value = 2
    return value 

return_random_sample()
```

b. Make a normalized histogram of the resulting values. Try a few values for the number of samples. Approximately how many samples do you require to get a qualitatively good (say, within 10%) agreement with the above probabilities?

```python
list_values = np.array([return_random_sample() for i in range(1000)])

plt.hist(list_values, density=True, bins=range(4))
plt.ylabel('P(x)')
plt.xlabel('x')
plt.show()
```

### 1.6 Conditional Probabilities

*Topics: frequentist probabilities, conditional probabilities*

Consider the probability table:

| X | Y | p(X,Y) |
|------|------| ----- |
| Hot (0)   | Sun (0) | 0.4 |
| Hot (0) | Rain (1) | 0.1 |
| Cold (1)  | Sun (0) | 0.2 |
| Cold (1)  | Rain (1) | 0.3 |



a. Implement a sampling algorithm that samples $(X, Y)$ pairs.

```python
def return_random_sample_joint():
    prob_value = np.random.random()
    x_value = 0
    y_value = 0
    if prob_value < 0.4:
        x_value = 0
        y_value =0
    elif prob_value < 0.5:
        x_value = 0
        y_value =1
    elif prob_value < 0.7:
        x_value=1
        y_value = 0
    elif prob_value <1:
        x_value =1
        y_value =1
    return [x_value,y_value]
```

b. Generate samples from $p(Y| X = 0)$ by removing ("rejecting") some samples generated by your above sampling algorithm. Use these to generate a histogram for the conditional distribution $p(Y| X=0)$.

```python
list_values_new=[]
for j in range(1000):
    if return_random_sample_joint() == [0,0]:
        list_values_new.append(0)
    if return_random_sample_joint() == [0,1]:
        list_values_new.append(1)

plt.hist(list_values_new, density=True, bins=range(3))
plt.ylabel('P(x) Marg')
plt.xlabel('x')
plt.show()
```

Mathematically, the PDF of a sum of two random variables is the convolution of their individual PDFs (do you understand why?). Let's call their PDFs $f(x)$ and $g(x)$, then the convolution is defined as

$$ (f \circledast g)(x) = \int_{-\infty}^{\infty} f(\tau) g(x - \tau) d\tau \, .$$

Program this convolution yourself for the two normal distributions $\mathcal{N}(x|\mu=2, \sigma=1)$ and $\mathcal{N}(x|\mu=4, \sigma=2)$ and show that you can recover the same connections seen already for the sets $\vec{x}$, $\vec{y}$ and $\vec{z}$.

**Hint**: You are **not** allowed to use a predefined convolution function. Approximate the integral by a sum over small elements with width $\Delta \tau$.

```python
def convolution(f, g, x, delta):
    N = len(x)
    conv = np.zeros(N)
    for i in range(N):
        for j in range(N):
            if i-j >= 0:  # Ensure non-negative index
                conv[i] += f[j] * g[i-j] * delta
    return conv

x = np.linspace(-10,10, 100)
delta = x[1]-x[0]
```

### 2.4 Marginal normal distributions

*Topics: multivariate normal & marginal distributions*

Consider a multivariate Gaussian of the form
$$P(\mathbf{x}) = \mathcal{N}(\mathbf{x}|\mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{(2\pi)^{D/2} |\mathbf{\Sigma}|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x}-\boldsymbol{\mu})^{T} \mathbf{\Sigma}^{-1}(\mathbf{x}-\boldsymbol{\mu})\right) \, ,$$

where $\mathbf{x}$ and $\boldsymbol{\mu}$ are vectors with $D$ entries and $\mathbf{\Sigma}$ is a $D \times D$ covariance matrix. $|\mathbf{\Sigma}|$ is the determinant of this matrix.

#### 2.4.1 (2 points)
Consider the case where $D = 2$. Use the following three sets of ($\boldsymbol{\mu}$, $\mathbf{\Sigma}$) to create 2D contour- and 3D surface plots of the corresponding PDFs (with arbitrary contour levels)

$$ \boldsymbol{\mu_1} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_1} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_2} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_2} = \begin{pmatrix} 1 & -0.5 \\ -0.5 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_3} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_3} = \begin{pmatrix} 1 & 0.7 \\ 0.7 & 1 \end{pmatrix} \, .$$

Implement the multivariate Gaussian yourself using numpy matrix multiplication.

```python
def mvgauss(x, mu, sigma,D):
    denominator  = (2 * np.pi)**(D/2) * np.sqrt(np.linalg.det(sigma))
    numerator  = np.exp(-1/2 * np.sum(np.matmul((x-mu), np.linalg.inv(sigma)) * (x-mu), axis =-1))
    #numerator = np.exp(-0.5 * np.sum(np.matmul((x - mu), np.linalg.inv(sigma)) * (x - mu), axis=-1))
    #numerator = np.exp(-0.5 * np.sum(np.matmul((x - mu), np.linalg.inv(sigma)) * (x - mu), axis=-1))
    f = numerator / denominator
    return f

mean = np.array([5, 5])

sigma_1 = np.array([[1,0] , [0,1]])
sigma_2 = np.array([[1,-0.5] , [-0.5,1]])
sigma_3 = np.array([[1,0.7] , [0.7,1]])

# Generate a grid of points
x, y = np.meshgrid(np.linspace(2, 8, 100), np.linspace(2, 8, 100))
grid = np.dstack((x, y))

distribution_1 = mvgauss(grid, mean, sigma_1,2)

distribution_2 = mvgauss(grid, mean, sigma_2,2)

distribution_3 = mvgauss(grid, mean, sigma_3,2)
```
